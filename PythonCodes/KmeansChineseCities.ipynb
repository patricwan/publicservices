```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans

import numpy as np 
import pandas as pd

dpath = 'data/'
dfCities = pd.read_csv(dpath + "ChineseCitiesLongLat.csv")

plt.scatter(dfCities['Longitude'], dfCities['Latitude'], marker='o')
plt.show()

X = np.mat([dfCities['Longitude'], dfCities['Latitude']])
X = X.T

from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=10).fit_predict(X)
X = np.array(X)


plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()

from sklearn import metrics
metrics.calinski_harabaz_score(X, y_pred)  
```

```python
from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=15).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()

from sklearn import metrics
metrics.calinski_harabaz_score(X, y_pred)  
```

```python
from sklearn.cluster import MiniBatchKMeans

yPreMini = MiniBatchKMeans(n_clusters = 10).fit(X)

pre_clu = yPreMini.labels_

plt.scatter(X[:, 0], X[:, 1], c=pre_clu)
plt.show()


from sklearn import metrics
metrics.calinski_harabaz_score(X, y_pred) 
```
