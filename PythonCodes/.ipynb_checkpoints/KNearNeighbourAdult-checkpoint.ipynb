```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn import  metrics
from sklearn.metrics import accuracy_score

import seaborn as sns
%matplotlib inline


dpath = './data/adult/'
train = pd.read_csv(dpath +"adultTrainLamda.csv")
train.head()

```

```python
test = pd.read_csv(dpath +"adultTestLamda.csv")
test.head()
```

```python
y_train = train['IncomeRes']    
x_train = train.drop(["IncomeRes"], axis=1)

y_test = test['IncomeRes']    
x_test = test.drop(["IncomeRes"], axis=1)
```

```python
from sklearn import neighbors

n_neighbors = 15

for weights in ['uniform', 'distance']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    scoreFloat = clf.fit(x_train, y_train).score(x_test, y_test)
    print("ScoreFloat",scoreFloat)



```
