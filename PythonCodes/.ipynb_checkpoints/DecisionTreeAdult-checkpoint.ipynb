```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn import  metrics
from sklearn.metrics import accuracy_score

import seaborn as sns
%matplotlib inline


dpath = './data/adult/'
train = pd.read_csv(dpath +"adultTrainClean.csv")
train.head()
```

```python
test = pd.read_csv(dpath +"adultTestClean.csv")
test.head()
```

```python
y_train = train['IncomeRes']   #形式为Class_x
x_train = train.drop(["IncomeRes"], axis=1)

y_test = test['IncomeRes']   #形式为Class_x
x_test = test.drop(["IncomeRes"], axis=1)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier().fit(x_train, y_train)
print("clf",clf)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("adultDecisionTree") 
```

```python
y_pred = clf.predict(x_test)

yArray = np.array(y_test)
print("Testing Score:%f" % (accuracy_score(yArray, y_pred)))
```
